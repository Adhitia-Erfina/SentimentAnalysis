```python
pd.set_option('display.max_columns', None)
data_clean = pd.read_csv('/content/data_clean.csv')
data_clean.head(10)
```

```python
#Visualisasi Sentimen Positif dan Negatif

plt.figure(figsize=(5, 3))
sns.countplot(data=data_clean, x='Label', palette={0: "red", 1:"lightskyblue"})
plt.xlabel('Sentimen')
plt.ylabel('Jumlah')
plt.title('Visualisasi Sentimen Positif dan Negatif')
plt.show()
```

```python
data_negatif = data_clean[data_clean['Label'] == 0]
data_positif = data_clean[data_clean['Label'] == 1]
```

```python
#Visualisasi Kata Negatif

all_text_s0 = ' '.join(word for word in data_negatif["Komentar"])
wordcloud = WordCloud(colormap='Reds', width=1000, height=1000, mode='RGBA', background_color='white').generate(all_text_s0)
plt.figure(figsize=(9, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Visualisasi Kata Negatif')
plt.margins(x=0, y=0)
plt.show()
```

```python
#Visualisasi Kata Positif
all_text_s0 = ' '.join(word for word in data_positif["Komentar"])
wordcloud = WordCloud(colormap='Blues', width=1000, height=1000, mode='RGBA', background_color='white').generate(all_text_s0)
plt.figure(figsize=(9, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Visualisasi Kata Positif')
plt.margins(x=0, y=0)
plt.show()
```
